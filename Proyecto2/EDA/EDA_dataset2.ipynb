```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
ruta = "C:/Users/mico1/OneDrive/Favorites/Documents/GitHub/Examen2/Proyecto2/datasets/ict.csv"
df = pd.read_csv(ruta,sep=";")
df
```

```python
df.info()
```

```python
df = df.rename(columns={'Unnamed: 0': 'Country'})
df.info()
```

```python
# Reemplazar comas por puntos y convertir a numérico
for col in df.columns[1:]:
    df[col] = df[col].str.replace(',', '.').str.replace('m', '').str.strip()
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df.describe()
```

```python
df = df.dropna(how='all')
df.info()
```

```python
unnamed_cols = ['Unnamed: 3', 'Unnamed: 4', 'Unnamed: 5', 'Unnamed: 6', 'Unnamed: 7']

for col in unnamed_cols:
    if col in df.columns:
        # Solo si la columna es numérica
        if pd.api.types.is_numeric_dtype(df[col]):
            mean_val = df[col].mean()
            df[col] = df[col].fillna(mean_val)
            print(f"Columna {col}: promedio usado para rellenar = {mean_val:.2f}")
        else:
            print(f"Columna {col} no es numérica, no se rellena.")
```

```python
df.info()
```

```python
print(df['ICT use at and for school'].unique())
```

```python
# Eliminar posibles outliers (por ejemplo, valores mayores a 100 si es porcentaje)
col = 'ICT use at and for school'

# Opcional: filtrar valores atípicos (por ejemplo, mayores a 100 si es porcentaje)
df[col] = df[col].apply(lambda x: np.nan if x is not np.nan and x > 100 else x)

# Calcular el promedio ignorando NaN
mean_val = df[col].mean()
print(f"Promedio de '{col}' (sin outliers): {mean_val:.2f}")

# Rellenar los NaN con el promedio
df[col] = df[col].fillna(mean_val)
```

```python
df.info()
```

```python
col = 'Students who reported the use of computers in mathematics lessons during the month prior to the PISA test'

# Opcional: eliminar outliers (por ejemplo, valores mayores a 100 si es porcentaje)
df[col] = df[col].apply(lambda x: np.nan if x is not np.nan and x > 100 else x)

# Calcular el promedio ignorando NaN
mean_val = df[col].mean()
print(f"Promedio de '{col}' (sin outliers): {mean_val:.2f}")

# Rellenar los NaN con el promedio
df[col] = df[col].fillna(mean_val)
```

```python


# Histograma
plt.figure(figsize=(12,5))
sns.histplot(df['ICT use at and for school'], bins=15, kde=True, color='skyblue')
plt.title('Histograma: Uso de TIC en la escuela (%)')
plt.xlabel('Porcentaje de uso de TIC')
plt.ylabel('Frecuencia')
plt.show()

# Boxplot
plt.figure(figsize=(8,4))
sns.boxplot(x=df['ICT use at and for school'], color='lightgreen')
plt.title('Boxplot: Uso de TIC en la escuela (%)')
plt.xlabel('Porcentaje de uso de TIC')
plt.show()
```

```python
# Selecciona solo las columnas numéricas (excluyendo 'Country')
numeric_cols = df.select_dtypes(include='number')

# Calcular la matriz de correlación
corr = numeric_cols.corr()

# Graficar el mapa de calor
plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Mapa de calor de correlación entre variables numéricas')
plt.show()
```
